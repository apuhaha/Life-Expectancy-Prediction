### Splitting the dataset into dependent and independant variables

```python
y = data['Life expectancy '].values
x = data.drop(['Life expectancy '], axis=1).values
```

### Splitting the dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 20)
```

### Modeling

```python
lin = LinearRegression()
```

```python
# fitting the model to the train data
lin.fit(x_train, y_train)
```

```python
y_pred = lin.predict(x_test)
```

### Evaluating the model

```python
from sklearn.metrics import mean_squared_error, accuracy_score
print('R^2: {}'.format(lin.score(x_test, y_test)))
print('rmse: {}'.format(np.sqrt(mean_squared_error(y_pred, y_test))))
```
